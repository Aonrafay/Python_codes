# AI Grade and Attendance validator

```python
# create AI Grade and attendance validator function
def validate_grade_attendance(grade, attendance):
    if grade >= 90 and attendance >= 95:
        return "Excellent"
    elif grade >= 80 and attendance >= 90:
        return "Good"
    elif grade >= 70 and attendance >= 85:
        return "Average"
    else:
        return "Needs Improvement"
# call the validate_grade_attendance function
s_marks=int(input("Enter your marks: "))
s_attendance=int(input("Enter your attendance percentage: "))
status=validate_grade_attendance(s_marks, s_attendance)
print(f"Your performance is: {status}")
```

# Smart Multiplication Table
>use of Functions

>use of Loops

```python
# generate the multiplication system for the start and end number
def mul_table(number ,limit):
    print(f"Multiplication Table for {number} up to {limit}:")
    for i in range(1, limit + 1):
        print(f"{number} x {i} = {number * i}")
    # user input for number and limit
num = int(input("Enter a number to generate its multiplication table: "))
limit = int(input("Enter the limit for the multiplication table: "))
mul_table(num, limit)
    
```
