# Part 6: Loops
## FOR Loop
When you know the count or have a specific list to go through.\
range(5): 0, 1, 2, 3, 4 \
range(1, 6): 1, 2, 3, 4, 5 (Stops before the last number!)
### Syntax

for item in sequence:\
    # Code to repeat for every item\
    print(item)

## While Loop
A While Loop is used when you want to repeat an action, but you don't know exactly how many times it will take. It keeps going as long as a specific rule is True.
### Syntax

while condition:\
    # Code to repeat\
    # IMPORTANT: You must change something here \
    # so the condition eventually becomes False!

````python
# Factorial using For Loop
n = int(input("Enter a number for Factorial: "))
fact = 1
for i in range(1, n + 1):
    fact = fact * i # fact *= i equls to  fact = fact * i
print(f"Factorial of {n} is: {fact}")
````

````python
# Factorial using While Loop
n = int(input("Enter a number: "))
fact = 1
temp_n = n  # We use a temporary variable to keep the original 'n' safe

while temp_n > 0:
    fact *= temp_n     # Multiply the current number into the result
    temp_n -= 1        # Decrease the number (The Update Rule)

print(f"The factorial of {n} is {fact}")
````

````python
# Count using While Loop
count = 5

while count > 0:
    print(f"Counting down: {count}")
    count -= 1  # This is the "Update Rule". It makes count smaller.

print("Blast off!")
````

## Scope of Variables
"Scope" refers to where a variable can be seen or used in your code.  
`Local Scope:` A variable created inside a function. It only exists while that function is running. It’s like a "secret" inside a room.  
`Global Scope:` A variable created outside all functions. It can be seen by everyone. It’s like a "sign" on the street.

````python
x = "I am Global" # Everyone can see this

def my_room():
    y = "I am Local" # Only people in this room can see this
    print(y)
    print(x) # The room can see the outside sign

my_room()
# print(y) # <--- This would cause an ERROR because 'y' doesn't exist out here!
````

````python
x = "I am Global" # Everyone can see this

def my_room(x):
    y = "I am Local" # Only people in this room can see this
    print(y)
    print(x) # The room can see the outside sign

my_room(x)
# print(y) # <--- This would cause an ERROR because 'y' doesn't exist out here!
````

## Task 8 The "Smart" Multiplication Table with Limit
Scenario: A student wants to see a table of a number, but they want to choose where the table stops (e.g., Table of 5 up to 15).   
Requirements:  
Use a for loop and function.  
Use range(start, stop + 1) to include the user's limit.   
Use f-strings for clean output.  

````python
def print_custom_table(number, limit):
    print(f"\n--- Table of {number} up to {limit} ---")
    for i in range(1, limit + 1):
        # Assignment operator logic: result = number * i
        print(f"{number} x {i} = {number * i}")

# User Inputs
num = int(input("Which table do you want? "))
stop = int(input("Where should the table stop? "))

print_custom_table(num, stop)
````

## Multiple Parameters Less Variables
If you have 3 parameters but only send 2 variables, Python will get upset! It’s like a recipe that requires Flour, Eggs, and Sugar, but you forget to bring the Sugar. The "cake" (the function) cannot be made.   
### The Error: "Missing Arguments"
If you don't provide enough information, Python throws a TypeError.   
```python 
def make_sandwich(bread, filling, sauce):
    print(f"Making a {filling} sandwich on {bread} with {sauce}!")

# ERROR: We only gave 2 things, but it needs 3!
make_sandwich("Brown Bread", "Chicken")

### The Solution: Default Parameters
You can give a parameter a Default Value. This makes that piece of information optional. If the user forgets to send it, Python uses the default value you set.

### The Rules of Default Parameters
Default parameters must come LAST.  
- Correct: def test(a, b, c="Default"):  
- Incorrect: def test(a="Default", b, c): (Python will crash before even starting).  

````python
# We set 'sauce' to "Mayo" by default
def make_sandwich(bread, filling, sauce="Mayo"):
    print(f"Making a {filling} sandwich on {bread} with {sauce}!")

# Now this works! It will use "Mayo" automatically.
make_sandwich("White Bread", "Egg")

# You can still change it if you want:
make_sandwich("Brown Bread", "Beef", "Ketchup")
````

# Part 8:Data Structures 
Data Structures are the "containers" we use to hold our datasets. Before we can train a model, we need to organize our data (like images, house prices, or names) into these structures so Python can process them quickly. Think of Data Structures as different types of Storage Boxes in a warehouse. Some boxes are easy to open and change, while others are locked shut for safety.

## 1. Lists: The Versatile Shopping Bag
Definition: A List is an ordered collection that is changeable (mutable) and allows duplicate members.  
Analogy: A shopping bag. You can add items, take them out, or swap a Pepsi for a Coke.  
Syntax: `my_list = [item1, item2, item3]`

````python
fruits = ["Apple", "Banana", "Cherry"]
fruits.append("Orange")     # Add to the end
fruits[1] = "Mango"         # Change 'Banana' to 'Mango'
print(fruits[0:4])          # Slicing: Get first two items
````

## 2.Tuples: The Locked Safe
Definition: A Tuple is an ordered collection that is unchangeable (immutable).  
Analogy: A birth certificate. Once written, you don't change the date or the name. We use these for data that must stay "safe."  
Syntax: `my_tuple = (item1, item2, item3)`

````python
dimensions = (1920, 1080)
# dimensions[0] = 800  <-- This would cause an ERROR!
print(f"Width is: {dimensions[0]}") 
dimensions[0:]
````

### How to `append()` in Tuples:
#### Trick A: The "New Statue" Method (Concatenation)   
You can add two tuples together to create a brand-new third tuple.   

````python
# Create a tuple
vowels = ("a", "e", "i")

# Add another tuple to it (must have a comma if it's a single item!)
vowels = vowels + ("o","u" )

print(vowels) # Output: ('a', 'e', 'i', 'o', 'u')                                      
````

#### Trick B: The "Temporary Bag" Method (Casting)
You turn the tuple into a list, change it, and turn it back.

````python
# 1. Start with a tuple
colors = ("Red", "Green")

# 2. Convert to a List (The Shopping Bag)
temp_list = list(colors)

# 3. Append to the list
temp_list.append("Blue")

# 4. Convert back to a Tuple (Lock the Safe)
colors = tuple(temp_list)

print(colors) # Output: ('Red', 'Green', 'Blue')
````

## 3.Sets: The Magic Bag
Definition: A Set is a collection that is unordered, unindexed, and has no duplicate members.  
Analogy: A bag of unique coins. If you try to put a second identical coin in, it simply disappears.  
Syntax: my_set = {item1, item2}

````python
ids = {101, 102, 103, 101} # 101 is a duplicate
print(ids) # Output: {101, 102, 103} (Duplicates are removed!)
````

If you want to `append` a single item into a set at once, you use the `add()` method.  
If you want to `append` a whole list of items into a set at once, you use the `update()` method.

````python
my_set = {1, 2, 3}

# Try to add a NEW number
my_set.add(4)
print(my_set) # Output: {1, 2, 3, 4}

# Try to add a DUPLICATE number
my_set.add(2) 
print(my_set) # Output: {1, 2, 3, 4} (It didn't change!)
````

````python
skills = {"Python", "ML"}
new_skills = ["DL", "NLP", "Python"] # Python is a duplicate here

skills.update(new_skills)
print(skills) # Output: {'Python', 'ML', 'DL', 'NLP'}
````

# Part 9: OOP

## Class vs. Object
Class: The Blueprint. It’s the design for a car on paper. It defines what a car has (color, engine) and what it does (drive, brake).   
Object: The Real Car. It is the physical version built from that blueprint. You can build 100 different cars (Objects) from one single blueprint (Class).
```PYTHON
class Student:
    # This is the Blueprint for a Student
    pass

# Creating an Object (Instance)
s1 = Student()

## The Constructor (__init__) and Destructor (__del__)
The Constructor (__init__)
This is a special function that runs automatically the moment you create an object. We use it to set the initial values (like a student's name).

self: This represents the specific object being created. Think of it as saying "My name is..." or "My age is...".

The Destructor (__del__)
This runs when an object is deleted or the program finishes. It’s like a "cleanup" crew.

````python
class Robot:
    def __init__(self, name):
        self.name = name
        print(f"Robot {self.name} is ONLINE.")

    def __del__(self):
        print(f"Robot {self.name} is shutting down.")

# Creating the object triggers __init__
my_bot = Robot("Robo-1")

# Deleting the object triggers __del__
del my_bot
````

````python
class Human:
    def __init__(self, n, o):
        self.name = n
        self.occupation = o

    def do_work(self):
        if self.occupation == "tennis player":
            print(self.name, "plays tennis")
        elif self.occupation == "actor":
            print(self.name, "shoots film")

    def speaks(self):
        print(self.name, "says how are you?")

ali = Human("ali","actor")
ali.do_work()
ali.speaks()

ahmed = Human("ahmed","tennis player")
ahmed.do_work()
ahmed.speaks()
````

**Constructor `(__init__)`:** Use it when the object needs to remember something (like its name, ID, or balance).  
**Regular Functions:** Use them when the object just needs to do something using external data.  
In **Machine Learning**, we almost always use a constructor to "initialize" the model's settings (like learning rate) before we call the "action" function   
`(like .fit() or .predict())`.

## Inheritance
Inheritance allows one class (Child) to take all the features of another class (Parent). This prevents us from writing the same code twice.

````python
# Parent Class
class AIModel:
    def __init__(self, name):
        self.name = name
    
    def info(self):
        print(f"This is a {self.name} model.")

# Child Class (Inherits from AIModel)
class Regressor(AIModel):
    def predict(self):
        print("Predicting continuous values...")

# Usage
model1 = Regressor("Linear Regression")
model1.info()    # Borrowed from Parent
model1.predict() # Own skill
````

## Task 15 
Create a parent class Device with a brand. Create a child class SmartPhone that adds ram.

````python
class Device:
    def __init__(self, brand):
        self.brand = brand

class SmartPhone(Device):
    def __init__(self, brand, ram):
        # Use super() to call the Parent's constructor
        super().__init__(brand)
        self.ram = ram

    def show_specs(self):
        print(f"Brand: {self.brand}, RAM: {self.ram}GB")

p1 = SmartPhone("Samsung", 12)
p1.show_specs()
````
